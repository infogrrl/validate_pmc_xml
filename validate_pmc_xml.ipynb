```python
import csv
from lxml import etree
import pandas as pd
from pathlib import Path
```

```python
# Path to file of XML to validate
xml_folder = "../21550417-10-02_errors"

# Sample xml doc
xml_doc = "../21550417-10-02_errors/21550417-10-02-1499.xml"
```

```python
parser = etree.XMLParser()
```

```python
e = ""
```

```python
# <!DOCTYPE article SYSTEM "http://dtd.nlm.nih.gov/1.1/journalpublishing.dtd">
```

```python
try:
    etree.parse(xml_doc, parser)
except etree.XMLSyntaxError as e:
    with open(Path(xml_doc).with_suffix(".csv"), "w") as csvfile:
        csv_writer = csv.writer(csvfile, delimiter=",")
        log = e.error_log
        for entry in log:
            csv_writer.writerow(str(entry).split(":", 6))
    pass
```
